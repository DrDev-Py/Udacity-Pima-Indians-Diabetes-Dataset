## Investigation Overview


> In this investigation I aim to find out the most correlated variables with diabetes outcome in patients. 


## Dataset Overview

>This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases, version was downloaded from Kaggle.

>The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('diabetes.csv')
```

```python
# formatting columns to standard
df = df.rename(columns =lambda x: x.lower())
df.rename(columns={'outcome':'diabetes','diabetespedigreefunction' : 'diabetes_pedigree_function', 'bloodpressure':'blood_pressure','skinthickness':'skin_thickness'}, inplace=True)
```

```python
# changing diabetes to categorical datatype
from pandas.api.types import CategoricalDtype

cat = CategoricalDtype(categories=[0,1], ordered=True)
df.diabetes = df.diabetes.astype(cat)
```

## 1. Distribution of Diabetes Status

> __34.9% of participants have diabetes__ while __65.1% do not have diabetes.__

```python
# Creating a function for setting the size of figures
def size(x,y):
    plt.figure(figsize=[x,y])

outcomes = df.diabetes.value_counts()
size(10, 8)
plt.pie(outcomes, labels=['No','Yes'], startangle=90, autopct='%1.1f%%', counterclock=False, wedgeprops={'width':0.4});

plt.title("Composition of People with or without Diabetes", weight='bold', fontsize='x-large');
```

## 2. Are people with higher BMI more susceptible to diabetes?

> * Those **without diabetes** had a **median bmi of about 30**  and an **_IQR range of 25-35_**
> * While those **with diabetes** had a **median bmi of about 35** and an **_IQR range of about 31-39_**


```python
plt.style.use('tableau-colorblind10')

g = sns.FacetGrid(data=df, col='diabetes', sharex=False, height=6, aspect=1.1)
g.map(sns.boxplot, 'bmi', order=[0,1]);

g.set(xlabel='BMI')

ticks = [x for x in range(10, 70, 5)]
labels = [f'{x}' for x in ticks]

plt.xticks(ticks, labels);

plt.subplots_adjust(top=0.9)
plt.suptitle('Distribution of BMI Across Diabetes Status', fontsize='x-large', weight='bold');
```

## 3. What is the relationship between the age and Diabetes Status of the participants?
 
> * Most people with diabetes were older than those without diabetes, 
> * Most diabetic participants are around 30 years and 40 years as indicated by the distribution in the violinplot, with a larger population around the 30 year mark
> * The __median age of diabetic particpants around 36 years__
> * The **median age of nondiabetic participants is around 27 years**


```python
size(10,8)
sns.violinplot(data=df, x='diabetes', y='age', alpha=0.6, inner='quartile');

plt.xlabel("Diabetes", fontsize='x-large', weight="4")
plt.ylabel("Age", fontsize='x-large', weight="4")

ticks = [0,1]
labels = ticks
plt.xticks(ticks, labels, weight='bold')

plt.title("Variation of Age with Diabetes Status", fontsize=18, weight=6);
```

## 4. What influence does Glucose and Insulin levels on Daibetes?
 
> **OBSERVATION**
> * People with diabetes have a __higher glucose level__ than those without.

> * Diabetic participants had a __higher insulin level__ than those without diabetes although there __seems to be a lot of overlap__ between the two categories.

```python
bins = [-np.inf, 18.4, 24, 29, 34, 39, np.inf ]
labels = ['underweight','healthy-weight','overwieight','class-1-obesity','class-2-obesity','class-3-obesity']
df['bmi_category'] = pd.cut(df.bmi, bins=bins, labels=labels)


size(15,8)

plt.subplot(1,2,1)
plt.title('Glucose Levels in Diabetic and Non-Diabetic Patients', weight='heavy')
sns.boxplot(data=df, x='diabetes', y='glucose')
plt.ylabel('Glucose', weight='bold')
plt.xticks([0, 1], ["No", "Yes"], weight='bold');
plt.xlabel('Diabetes', weight='bold', fontsize=13)

plt.subplot(1,2,2)
plt.title('Insulin Levels in Diabetic and Non-Diabetic Patients', weight='heavy')
sns.boxplot(data=df, x='diabetes', y='insulin');
plt.ylabel('Insulin', weight='bold')
plt.xticks([0, 1], ["No", "Yes"], weight='bold');
plt.xlabel('Diabetes', weight='bold', fontsize=13);
```

## 5.What is the Relationship between the Ages, BMI Categories and Diabetes Status of the Participants?
 
> * There's a significant difference in the ages of diabetic and non-diabetic participants, 
> * There's no diabetic underweight participant.
> * There's a slight decrease in the age of diabetic participants as you move from healthy weight category down to the class-3-obesity category.

```python
plt.style.use('bmh')
size(14,8)
sns.pointplot(data = df, x = 'bmi_category', y = 'age', hue = 'diabetes',
                  dodge = 0.3, linestyles = "dotted", ci=95)

# plt.style.available

plt.title('Age Distribution of Different BMI Categories showing their Diabetes Status (CI=95%)', weight='heavy')
plt.ylabel('Age', weight='bold', size='xx-large', fontstyle='italic')
plt.xlabel('BMI Category', weight='bold', size='xx-large', fontstyle='italic')


ticks=np.arange(0,6,1)
plt.xticks(size='x-large', weight='light');
```

## 6. What is the Relationship between the Ages, BMI Categories and Diabetes Status of the Participants?
 
> * Glucose levels had the highest positive correlation with diabetes outcome, with a Pearson's Correlation Coefficient of 0.49

```python
size(13, 9)

# making a list of variables to be plotted on heatmap
df.diabetes = df.diabetes.astype(int)
corr = ['glucose','insulin','bmi', 'diabetes']

# correlation plo5
sns.heatmap(df[corr].corr(), annot = True,fmt = '.2f', cmap = 'Spectral_r', center=0, cbar_kws = {'label' : 'Correlation Coefficient'})
plt.title('Correlation Between Insulin, Glucose, BMI and Diabetes Status');
```
