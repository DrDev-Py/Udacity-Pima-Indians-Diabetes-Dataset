### 3. BMI Category

#### BMI  as a Diagnosing Tool for Body Fatness.

> * BMI are grouped into categories to vaguely define a person body fatness and what health risks they might be at risk for.
> * As such I would be categorising the BMI of the individuals into groups for a better idea of their body fatness.

> * A new column  using the [CDC Guidelines](https://www.cdc.gov/obesity/basics/adult-defining.html) would be used to classify the BMI column into ordered categories.

```python
#  Using pd.cut to group the bmi into categories

bins = [-np.inf, 18.4, 24, 29, 34, 39, np.inf ]
labels = ['underweight','healthy-weight','overwieight','class-1-obesity','class-2-obesity','class-3-obesity']
df['bmi_category'] = pd.cut(df.bmi, bins=bins, labels=labels)
```

### 6. Insulin

```python

sns.histplot(df, x=df.insulin);

# ticks = np.arange(0, df.insulin.max()+5, 30)
# labels = [f'{x}' for x in ticks]
# plt.xticks(ticks, labels, rotation =20);
```

> Having a lot of zero values which is not possible after a 2-hour serum-insulin test.

```python
zero_values = (df[df.insulin==0]['insulin']).count()

print(f'{zero_values}, \nPercentage composition of zero values: {((zero_values/df.shape[0])*100)}')
```

> About 48% of the data is zero, the __zero values__ will be __replaced__ using the __ffil() *forward fill* method__
> Here's why:
> * Replacing with mean or median would automatically create a mode of the value, since we would be replacing 48% of the whole column.
> * Using the __ffil__ method uses different values in respect to the position of the zero value and is appropriate since they're scattered all over the column, attributing a measure of randomness to the method.

```python
# Replacing the zero values with null so as to use the ffill() function
df.insulin.replace(0, np.nan, inplace=True)

df.insulin.fillna(method='ffill', inplace=True)
```

```python
#  Reploting the Histogram.
size(9,7)

sns.histplot(df, x=df.insulin);
plt.title('Distribution of Insulin Concentration after Insulin Test');

plt.xlabel("Insulin(mu U/ml)");
```

> * The distribution is highly skewed to the right, as such a logarithm transformation would be appropriate.

```python
size(9,7)

log_binsize = 0.05
bins = 10** np.arange(1, np.log10(df.insulin.max())+log_binsize, log_binsize)

sns.histplot(df, x=df.insulin, bins=bins);
plt.xscale('log')

ticks = [10, 20, 30, 50, 100, 200, 300, 500, 1000]
labels = [f'{x}' for x in ticks]

plt.xticks(ticks, labels);
plt.xlabel('INSULIN');
```

> **OBSERVATION**

> The Log distribution appears normal, also being multimodal, with 2 significant peaks betweem 50 and 100 and another 2 between 100 and 200, theres also another 2 betweeen 300 and 500.
